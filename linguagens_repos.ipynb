## Paginando os Repositrios

```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # Username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'
url
```

```python
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}'
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

### Exerccio

Durante essa etapa, devemos nos atentar a alguns detalhes:

criar uma lista vazia chamada followers_list para armazenar as informaes sobre os seguidores;

utilizar um lao para iterar e extrair os dados dos seguidores existentes em cada uma das pginas (dica: caso voc no tenha conhecimento de quantas pginas com seguidores existem, pesquise como utilizar o lao While True do Python);

acrescentar o parmetro "page" na url, alterando o valor dele a cada iterao do lao que estiver sendo usado para paginao (lembre-se que a numerao das pginas sempre comeam a partir do nmero 1);

enviar o token e a verso da API para o parmetro headers no momento de fazer a requisio GET.

```python
import requests

username = 'amzn'
url = f"https://api.github.com/users/{username}/followers"

# definindo o token e a verso da API
access_token = ' ' #removido
headers = {'Authorization': 'Bearer ' + access_token,
           'X-GitHub-Api-Version': '2022-11-28'}

page = 1
followers_list = []

while True: 

    # faz a requisio 
    url_page = f'{url}?page={page}'
    response = requests.get(url_page, headers=headers)

    # converte a resposta para um objeto JSON
    followers = response.json()

    # caso a lista esteja vazia, podemos sair do lao pois todos os dados foram extraidos
    if len(followers)==0:
        break

    # adicionando os seguidores a lista
    followers_list.append(followers)

    # incrementa o valor de 'page' para a prxima requisio
    page += 1
```
